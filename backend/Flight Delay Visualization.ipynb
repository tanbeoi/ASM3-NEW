DELAYED FLIGHTS VISUALIZATION:


```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
import pandas as pd

# Load the dataset
df = pd.read_csv('./BITRE_MERGED_FLIGHT_DELAY.csv')

# Display the first few rows of the dataset to understand its structure
print(df)
```

```python

# Create scatter plot
plt.figure(figsize=(15, 30))

# Scatter plot for Departures Delayed
sns.scatterplot(x='Airline', y='Departures Delayed', data=df, color='green', label='Departures Delayed')

# Scatter plot for Arrivals Delayed
sns.scatterplot(x='Airline', y='Arrivals Delayed', data=df, color='red', label='Arrivals Delayed')

# Customize the plot, fix x axis rotation labelling for ease of readability
plt.xticks(rotation=45, ha='right')
# title of visualization
plt.title('Scatter Plot of Flight Delays by Airline')
# Y Axis title
plt.ylabel('Frequency of Delays')
# X axis title
plt.xlabel('Airline')
plt.legend()
plt.tight_layout()

# Display Visualization
plt.show()
```

```python
# Line Plot Visualization indicating the flights delayed over the years
# we start by Grouping the data by year and summing up the delays
delays_over_years = df.groupby('Year')[['Departures Delayed', 'Arrivals Delayed']].sum().reset_index()

# Create a line plot 
plt.figure(figsize=(30, 15))

# Plot Delayed Departures
plt.plot(delays_over_years['Year'], delays_over_years['Departures Delayed'], marker='o', color='blue', label='Departures Delayed')

# Plot Delayed Arrivals
plt.plot(delays_over_years['Year'], delays_over_years['Arrivals Delayed'], marker='o', color='green', label='Arrivals Delayed')

# X label
plt.xlabel('Year')
# Y label
plt.ylabel('Number of Delays')
# Title of Visualization
plt.title('Flight Delays Over the Years')
plt.legend()
plt.grid(True)

# Display Visualization
plt.tight_layout()
plt.show()
```
